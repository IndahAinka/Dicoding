<h3>Explorasi Data Orders</h3>

```python
delivery_time = orders_df["delivery_date"] - orders_df["order_date"]
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
orders_df["delivery_time"] = round(delivery_time/86400)
```

```python
orders_df.describe(include="all")
```

<h3>Eksplorasi Data orders_df dan customers_df</h3>

```python
# Mengidentifikasi pelanggan yang belum pernah melakukan order, dengan menambahkan kolom status, active=pernah berlangganan, non-active= belum pernah berlangganan
customer_id_in_orders_df =  orders_df.customer_id.tolist()
customers_df["status"] = customers_df["customer_id"].apply(lambda x: "Active" if x in customer_id_in_orders_df else "Non Active")
customers_df.sample(5)
```

```python
# Melihat jumlah pelangggan yang aktif dan non aktif
customers_df.groupby(by="status").customer_id.count()
```

```python
# Melakukan left join
orders_customers_df = pd.merge(
    left=orders_df,
    right=customers_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
orders_customers_df.head()
```

```python
# Jumlah orders berdasarkan kota
orders_customers_df.groupby(by="city").order_id.nunique().sort_values(ascending=False).reset_index().head(10)
```

```python
# Jumlah orders berdasarkan state
orders_customers_df.groupby(by="state").order_id.nunique().sort_values(ascending=False)
```

```python
# Jumlah order berdasarkan gender
orders_customers_df.groupby(by='gender').order_id.nunique().sort_values(ascending=False)
```

```python
#Mendefinisikan sebuah kolom baru bernama “age_group”. 
#Kolom ini akan membantu kita dalam mengelompokkan pelanggan ke dalam tiga kelompok, yaitu youth, adults, dan seniors. 

orders_customers_df["age_group"] = orders_customers_df.age.apply(lambda x: "Youth" if x <= 24 else ("Seniors" if x > 64 else "Adults"))
orders_customers_df.groupby(by="age_group").order_id.nunique().sort_values(ascending=False)
```

```python
# Melihat produk yang paling laris
sales_product_df= pd.merge(
    left=sales_df,
    right=product_df,
    how='left',
    left_on="product_id",
    right_on="product_id"
)

sales_product_df.head()
```

```python
#Melihat informasi penjualan produk berdasarkan type produk
sales_product_df.groupby(by="product_type").agg({
    "sales_id": "nunique",
    "quantity_x": "sum",
    "total_price": "sum"
})
```

```python
# Melihat informasi produk berdasarkan nama produk
sales_product_df.groupby(by='product_name').agg({
    'sales_id': 'nunique',
    'quantity_x': 'sum',
    'total_price': 'sum'
}).sort_values(by='total_price', ascending=False)
```

<h3>Eksplorasi Data all_df</h3><br>

<body>Menampung semua informasi dari ke-empat table, untuk melihat pola demografi pelanggan</body>

```python
all_df= pd.merge(
    left= sales_product_df,
    right= orders_customers_df,
    how='left',
    left_on='order_id',
    right_on='order_id'
)

all_df.head()
```

```python
# Melihat preferensi pelanggan berdasarkan state
all_df.groupby(by=['state', 'product_type']).agg({
    'quantity_x': 'sum',
    'total_price': 'sum'
})
```

```python
# Melihat selera tipe produk pelanggan berdasarkan gender dan kelompok usia
all_df.groupby(by=['gender','product_type']).agg({
    'quantity_x': 'nunique',
    'total_price': 'sum'
})
```

```python
# Melihat selere tipe produc pelanggan berdasarkan umur pelanggan
all_df.groupby(by=['age_group','product_type']).agg({
    'quantity_x': 'nunique',
    'total_price': 'sum'
})
```

<h3>Bagaimana Performa Penjualan dan Revenue Perusahaan dalam Beberapa Bulan Terakhir?</h3>

```python
monthly_orders_df = all_df.resample(rule='M', on='order_date').agg({
    "order_id": "nunique",
    "total_price": "sum"
})
monthly_orders_df.index = monthly_orders_df.index.strftime('%Y-%m')
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "order_id": "order_count",
    "total_price": "revenue"
}, inplace=True)
monthly_orders_df.head()
```

```python
#Membuat Visualisasi Pertanyaan 1
monthly_orders_df = all_df.resample(rule='M', on='order_date').agg({
    "order_id": "nunique",
    "total_price": "sum"
})
monthly_orders_df.index = monthly_orders_df.index.strftime('%B') #mengubah format order date menjadi nama bulan
 
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "order_id": "order_count",
    "total_price": "revenue"
}, inplace=True)
 
plt.figure(figsize=(10, 5)) 
plt.plot(monthly_orders_df["order_date"], monthly_orders_df["order_count"], marker='o', linewidth=2, color="blue") 
plt.title("Number of Orders per Month (2021)", loc="center", fontsize=20) 
plt.xticks(fontsize=10) 
plt.yticks(fontsize=10) 
plt.show()
```
